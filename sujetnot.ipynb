```python
import numpy as np
import matplotlib. pyplot as plt
```

```python
#Question 3
L = 10
delta = 0.1
N = int(L//delta)
tf = 10
dt = 0.1

def T(N,delta):
    T = [[0 for i in range(N)]for j in range(N)]
    for x in range(N):
        for y in range(N):
            T[y][x] = 1/2 +np.sin(2*np.pi*x*delta)*np.sin(2*np.pi*y*delta)+1/2*np.cos(2*np.pi*x*delta)*np.cos(2*np.pi*y*delta)
    return np.array(T)

def derive(T,t):
    Derive = [[0 for i in range(N)] for j in range(N)]
    for i in range(1,N-1):
        for j in range(1,N-1):
            Derive[i][j] = -(T[i+1][j] - 4*T[i][j] + T[i-1][j] + T[i][j+1] + T[i][j-1])/(delta**2)
    return np.array(Derive)

def Euler_explicit(T0,delta,N,L,dt,tf):
    Temps = [0]
    t = 0
    Evolution = [T0]
    T = T0
    nombre_boucle_temps = int(tf//dt)
    
    for i in range(nombre_boucle_temps):
        t = t + dt
        Temps.append(t)
        T = T + dt*derive(T,t)
        Evolution.append(T)
        
    '''T = T + (tf-t)*derive(T,(tf-t))
    Evolution.append(T)
    Temps.append(tf)'''
    
    return Evolution,Temps

T0 = T(N,delta)
Evol,Temps = Euler_explicit(T0,delta,N,L,dt,tf)
#y = int(L//(3*delta))

    #Selection = [T[y]]
    #plt.imshow(T0)
    
fig,axes = plt.subplots(7,7)
for i in range(7):
    for j in range(7):
        axes[i,j].imshow(Evol[i+j])

#from matplotlib import animation
#anim=animation.FuncAnimation(fig, func = Evol, interval = 1000)
```
